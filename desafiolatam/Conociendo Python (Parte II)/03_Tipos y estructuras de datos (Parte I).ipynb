## <font color='green'>Listas</font>


### <font color='green'>Creación y acceso</font>

````python
# Creación de la lista con frutas favoritas
frutas_favoritas = ['manzana', 'banana', 'uva', 'fresa']

# Preguntar al usuario por una fruta y agregarla a la lista
nueva_fruta = input("Agrega una fruta a la lista: ")
frutas_favoritas.append(nueva_fruta)

# Mostrar la lista completa
print("Lista de frutas favoritas:")
for fruta in frutas_favoritas:
    print("-", fruta)
````

### <font color='green'>Filtrado</font>

````python
# Crear una lista con los nombres de tus amigos
amigos = ['Ana', 'Carlos', 'Luis', 'María', 'Pedro', 'Patricia']

# Pedir al usuario que ingrese una letra
letra = input("Ingresa una letra: ")

# Filtrar los nombres que comienzan con la letra ingresada. Notar que se pasa a minúscula
filtrados = [amigo for amigo in amigos if amigo.lower().startswith(letra.lower())]

# Mostrar los nombres filtrados
print("Nombres que comienzan con la letra", letra + ":")
for amigo in filtrados:
    print("-", amigo)
````

### <font color='green'>Manipulación de listas</font>

````python
# Crea una lista con tus películas favoritas
peliculas_favoritas = ["El Padrino", "Titanic", "El Señor de los Anillos", "La Cenicienta"]

# Pide al usuario que agregue una película nueva a la lista
nueva_pelicula = input("Por favor, escribe el nombre de una nueva película que quieras agregar a la lista: ")
peliculas_favoritas.append(nueva_pelicula)

# Pide al usuario que elimine una película existente de la lista
pelicula_a_eliminar = input("Por favor, escribe el nombre de una película que quieras eliminar de la lista: ")
peliculas_favoritas.remove(pelicula_a_eliminar)

# Muestra la lista completa actualizada
print("Esta es tu lista de películas favoritas actualizada:")
for pelicula in peliculas_favoritas:
    print("-", pelicula)
````

### <font color='green'>Ordenando una lista</font>

````python
# Ordenar una lista
libros_favoritos = ["1984", "El Gran Gatsby", "Cien años de soledad", "Orgullo y prejuicio"]

# Ordena la lista alfabéticamente
libros_favoritos.sort()

# Muestra los libros en orden ascendente
print("Esta es tu lista de libros favoritos en orden alfabético:")
for libro in libros_favoritos:
    print(libro)
````

## <font color='green'>Diccionarios</font>


### <font color='green'>Creación y acceso</font>

````python
# Creación
mi_diccionario = {'nombre': 'Juan', 'edad': 30, 'ciudad': 'Lima'}
print(mi_diccionario['nombre'])  # salida: 'Juan'
print(mi_diccionario['edad'])  # salida: 30

# Crea un diccionario con los nombres de tus amigos y la ciudad donde viven
amigos = {"Juan": "Bogotá", "Maria": "Medellín", "Luis": "Cali", "Ana": "Barranquilla"}

# Pide al usuario que agregue un amigo nuevo y la ciudad donde vive
nuevo_amigo = input("Por favor, escribe el nombre de un nuevo amigo: ")
ciudad_nuevo_amigo = input("Por favor, escribe la ciudad donde vive el nuevo amigo: ")
amigos[nuevo_amigo] = ciudad_nuevo_amigo

# Muestra el diccionario completo
print("Este es tu diccionario de amigos y ciudades:")
for amigo, ciudad in amigos.items():
    print(f"- {amigo} vive en {ciudad}")
````

### <font color='green'>Filtrado</font>

````python
# Crea un diccionario con los nombres de tus amigos y su edad
amigos_edades = {"Juan": 25, "Maria": 30, "Luis": 25, "Ana": 27}

# Pide al usuario que ingrese una edad
edad_buscar = int(input("Por favor, escribe una edad para buscar amigos que tengan esa edad: "))

# Muestra los nombres de los amigos que tienen la edad buscada
print(f"Amigos que tienen {edad_buscar} años:")
for amigo, edad in amigos_edades.items():
    if edad == edad_buscar:
        print("-", amigo)
````

### <font color='green'>Manipulación</font>

````python
# Crea un diccionario con los nombres de tus películas favoritas y su año de lanzamiento
peliculas = {"El Padrino": 1972, "Pulp Fiction": 1994, "El Rey León": 1994, "Titanic": 1997}

# Pide al usuario que agregue una película nueva con su año de lanzamiento
nueva_pelicula = input("Por favor, escribe el nombre de una nueva película: ")
anio_nueva_pelicula = int(input("Por favor, escribe el año de lanzamiento de la nueva película: "))
peliculas[nueva_pelicula] = anio_nueva_pelicula

# Pide al usuario que elimine una película existente
pelicula_eliminar = input("Por favor, escribe el nombre de una película para eliminarla: ")
del peliculas[pelicula_eliminar]

# Muestra el diccionario completo actualizado
print("Este es tu diccionario de películas actualizado y años de lanzamiento:")
for pelicula, anio in peliculas.items():
    print(f"- {pelicula} se lanzó en {anio}")
````

### <font color='green'>Orden</font>

````python
# Crea un diccionario con los nombres de tus libros favoritos y su autor
libros = {"El Quijote": "Miguel de Cervantes", "Cien años de soledad": "Gabriel García Márquez", "La Iliada": "Homero", "Orgullo y prejuicio": "Jane Austen"}

# Ordena el diccionario alfabéticamente por el nombre del libro
libros_ordenados = dict(sorted(libros.items()))

# Muestra los libros en orden ascendente
print("Estos son tus libros favoritos ordenados alfabéticamente:")
for libro, autor in libros_ordenados.items():
    print(f"- {libro}, escrito por {autor}")
````

### <font color='green'>Listas y diccionarios</font>

Para transformar de listas a diccionarios en Python, puedes usar la función **zip()**, que combina dos listas en pares de valores. Luego deberás convertir el objeto zip en una **lista** o un **diccionario** utilizando las funciones **list()** o **dict()**, respectivamente. Por ejemplo:

````python
nombres = ["Juan", "María", "Pedro"]
edades = [28, 30, 25]

diccionario_edades = dict(zip(nombres, edades))

print(diccionario_edades)
````

Para transformar de **diccionarios** a **listas**, puedes usar los métodos **keys()** y **values()**.
Por ejemplo:

````python
diccionario_edades = {'Juan': 28, 'María': 30, 'Pedro': 25}

nombres = list(diccionario_edades.keys())
edades = list(diccionario_edades.values())

print(nombres)
print(edades)
````

Como vemos en estos ejemplos, para transformar de **diccionario** a **lista** obtendremos o las llaves o los valores del diccionario.

Por otro lado para transformar de **lista** a **diccionario** debemos pasarle pares llave-valor o **tuplas** para adaptar la estructura a la de un **diccionario.**

## <font color='green'>Tuplas</font>

Las tuplas en Python son estructuras de datos similares a las listas, pero a diferencia de ellas son **inmutables,** es decir, una vez creadas no se pueden modificar.

Para crear una tupla en Python se utiliza una sintaxis similar a la de las listas, pero en lugar de corchetes, se utilizan paréntesis.

````python
mi_tupla=(1,2,3,4)
print(mi_tupla[3])
````

````python
# ¡Aquí tendremos un error!
mi_tupla[1]=6
print(mi_tupla)
````

## <font color='green'>Sets</font>

Los **sets** en Python son una estructura de datos que permite almacenar elementos únicos e inmutables en un conjunto sin orden. Una gran ventaja de los sets es **la eliminación de elementos duplicados de una lista.**

````python
my_set = set([1, 2, 3, 3, 3, 3])
print(my_set)
````

### <font color='green'>Filtrado</font>

````python
import numpy as np

lista = [1, 6, 3, 8, 2, 9,10]

arr = np.array(lista)

arr[arr > 5]
````

Para filtrar un array por más de una condición, se puede combinar las condiciones utilizando las funciones:
- **numpy.logical_and().**
- **numpy.logical_or().**

Por ejemplo, si tenemos un arreglo de números aleatorios y queremos filtrar los valores que sean mayores a 5 y menores a 10, podemos hacer lo siguiente:

````python
arr[np.logical_and(arr > 5, arr < 10)]
````

### <font color='green'>Otros métodos</font>

- **np.zeros():** Crea un array de ceros con la forma especificada.
- **np.ones():** Crea un array de unos con la forma especificada.
- **np.arange():** Crea un array con valores espaciados uniformemente en un intervalo especificado.
- **np.linspace():** Crea un array con valores espaciados uniformemente en un rango especificado.

````python
a=np.zeros(3)
b=np.ones(3)
c=np.arange(2,30,5)#inicio, fin, distancia entre valores
d=np.linspace(2,3,5)#inicio, fin, cantidad de valores
print(a)
print(b)
print(c)
print(d)
````

## <font color='green'>Manejo de datos con NumPy Arrays</font>

### <font color='green'>Carga de datos con loadtxt</font>

````python
nombres = np.loadtxt("nombre_astronautas.csv", delimiter=',', dtype=str, encoding='utf-8')
````

Los argumentos que se pasan a la función son los siguientes:

- **nombre_astronautas.csv:** este es el nombre del archivo que se desea cargar.

- **delimiter:** este argumento indica que el delimitador utilizado en el archivo es la coma (','). Es decir, cada valor en el archivo está separado por comas.

- **dtype:** este argumento indica que se desea cargar los datos como cadenas de texto (strings).

- **encoding:** este argumento indica que el archivo está codificado en UTF-8. UTF-8 es un tipo de codificación de caracteres que permite representar la mayoría de los caracteres utilizados en el mundo.

Adiiconalmente, podemos utilizar alguns otros parámetros optativos que nos serán útiles para la carga de archivos separados por coma, como los **csv*:

- **skiprows:** Con este parámetro puedes indicar el número de filas que quieres saltar al leer el archivo. Esto es útil si el archivo tiene filas de encabezado que no contienen datos, por ejemplo.

- **usecols:** Este parámetro te permite seleccionar las columnas que quieres leer del archivo. Puedes pasarle un número de columna (empezando por cero) o una lista de números si necesitas leer varias columnas.

Por ejemplo, supongamos que tenemos un archivo de datos separados por comas:
a,	b,	c,	d<br>
1,	4,	6,	2<br>
2,	5,	4,	1<br>
3,	4,	9,	7<br>
4,	8,	0,	3<br>
5,	6,	9,	9<br>
6,	23,	87,	3<br>
7,	8,	6,	75<br>
8,	5,	4,	53<br>
9,	9,	2,	43<br>
10,	6,	2,	3<br>

Utilizamos los parámetros del método loadtxt() para indicar que el archivo tiene separación por comas, también indicamos que solo considere las columnas 1 y 2,  y también que se salte la fila que tiene los encabezados.


````python
datos = np.loadtxt('datos.csv', delimiter=',', skiprows=1, usecols=(1, 2), dtype=int)
print(datos)
````

### <font color='green'>Estadísticas con NumPy Arrays</font>

````python
valores = np.array([5,2,7,3,7,4,8,4,2,5,8,4,7,4,7,9,0,5,3,3])
print(np.mean(valores))#Calcula la media de los elementos del array.
print(np.median(valores))#Calcula la mediana de los elementos del array.
print(np.std(valores))#Calcula la desviación estándar de los elementos del array.
print(np.var(valores))#Calcula la varianza de los elementos del array.
print(np.percentile(valores,75))#Calcula el percentil 75 de los elementos del array.
````
